```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=100, n_features=2,  n_informative=1 ,n_redundant=0,
n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
plt.scatter(X[:,0],X[:,-1],c=y ,s= 100)
```

```python
def step_function(z):
    return 1 if z >0 else 0
```

```python
class Logist:

    def __init__(self,lr,epoches):
        self.epoches = epoches
        self.lr = lr
        self.weights = None
    
    def fit(self,X,y):

        new_X = np.insert(X,0,1,axis=1)
        self.weights = np.ones(new_X.shape[1])
        # print(new_X)

        for i in range(self.epoches):
            ran = np.random.randint(0,X.shape[0])
            y_hat = step_function(np.dot(new_X[ran],self.weights))
            self.weights = self.weights + self.lr * (y[ran] - y_hat) * new_X[ran]


        plt.scatter(X[:,0],X[:,-1],c= y,s=100)
        m = -(self.weights[1]/self.weights[2])
        b = -(self.weights[0]/self.weights[2])


        X_input = np.linspace(-3,3,100)
        y_input = m*X_input + b


        plt.plot(X_input,y_input,color='red',linewidth = 3)
        plt.scatter(X[:,0],X[:,1],c = y,cmap='RdYlGn',s =100)
        plt.ylim(-3,2)

    def predict(self,X_test):
        if X_test.shape[0] > 1:
           X_test = np.insert(X,0,1,axis=1)
           y_pred = np.dot(X_test, self.weights)

           for i in range(X_test.shape[0]):
              print(step_function(y_pred[i]))
```

```python
log = Logist(0.2,110)
log.fit(X,y)
pred = log.predict(X)
```
