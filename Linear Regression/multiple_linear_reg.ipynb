```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import numpy as np
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import mean_squared_error,r2_score
```

```python
X,y = make_regression(n_samples=200,n_features=2,noise=40)
```

```python
data = pd.DataFrame({'feature 1': X[:,0], 'feature 2': X[:,-1],'target':y})
```

```python
px.scatter_3d(data,x=data['feature 1'],y=data['feature 2'],z=data['target'])
```

```python
X_test,X_train,y_test,y_train = tts(X,y,train_size=0.2)
```

```python
class Multi_Reg:
    def __init__(self):
        self.intercept = 0
        self.coef = 0
    
    def fit(self,X_train,y_train):
        X_train = np.insert(X_train,0,1,axis=1)
        betas = np.linalg.inv(np.dot(X_train.T,X_train)).dot(X_train.T).dot(y_train)
        self.intercept = betas[0]
        self.coef = betas[1:]
        
        
    
    def predict(self,X_test):
        
            y_pred = np.dot(X_test,self.coef) + self.intercept
            return y_pred
        
       
    
    def score(self,y_test,y_pred,mse = False):
        print("Score ",r2_score(y_test,y_pred))
        if mse == True:
            print("MSE ", mean_squared_error(y_test,y_pred))

        
    
    
    
```

```python
reg = Multi_Reg()
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
reg.score(y_test,y_pred,mse=True)
```
